```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("output.csv")
traces = list(df.groupby(df.particle))
```

```python
for trace in traces[1:]:
    plt.figure(trace[0])
    plt.plot(traces[0][1][" px"], traces[0][1][" py"], label = "p%s" % traces[0][0])
    plt.plot(trace[1][" px"], trace[1][" py"], label = "p%s" % trace[0])
    plt.legend()
    plt.show()
```
