```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("output.csv")
traces = list(df.groupby(df.particle))
```

```python
plt.figure(1, figsize=(20,12))
ax = plt.axes(projection='3d')

zero = lambda k: [0 for _ in k]

for i in list(range(len(traces)))[::-1]:
    pnt_x = list(traces[i][1][" px"])
    pnt_y = list(traces[i][1][" py"])
    pnt_z = list(traces[i][1][" pz"])
    ax.plot3D(pnt_x, pnt_y, pnt_z, label = "p%s (%s, %s, %s)" % (traces[i][0], pnt_x[0], pnt_y[0], pnt_z[0]))

plt.legend()
plt.show()
```
